```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
np.random.seed(4)
m = 60
w1, w2 = 0.1, 0.3
noise = 0.1

angles = np.random.rand(m) * 3 * np.pi / 2 - 0.5
X = np.empty((m, 5))
X[:, 0] = np.cos(angles) + np.sin(angles)/2 + noise * np.random.randn(m) / 2
X[:, 1] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 2] = X[:, 1] * w1 + X[:, 0] * w2 + noise * np.random.randn(m)
X[:, 3] = np.sin(angles) * 0.7 + noise * np.random.randn(m) / 2
X[:, 4] = X[:, 0] * w2 + X[:, 2] * w1 + noise * np.random.randn(m)
```

```python
X_centered = X - X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:, 0]
c2 = Vt.T[:, 1]
X
```

```python
pca = PCA()             # 初始化一个 PCA 对象，不限制主成分数
pca.fit(X)              # 对数据 X 做 PCA 拟合
cumsum = np.cumsum(pca.explained_variance_ratio_)   # 计算累计解释方差比例
d = np.argmax(cumsum >= 0.95) + 1                   # 找到累计方差首次达到95%的主成分数量
d
```

```python
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X)
X_reduced
```
