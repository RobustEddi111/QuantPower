```python
import instaloader

# 创建实例
L = instaloader.Instaloader()

# 目标用户名
username = "henryss8s"

# 获取用户 profile
profile = instaloader.Profile.from_username(L.context, username)

# 限制下载数量（如只下载前5个帖子）
max_count = 5
count = 0

for post in profile.get_posts():
    if count >= max_count:
        break
    print(f"Downloading post {count + 1} - {post.url}")
    L.download_post(post, target=username)
    count += 1
```
