```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")
import os
import pandas as pd

def analyze_one_file(file_path, output_path):
    ext = os.path.splitext(file_path)[1].lower()
    try:
        if ext == ".xls":
            df = pd.read_excel(file_path, engine="xlrd")
        elif ext == ".xlsx":
            try:
                df = pd.read_excel(file_path, engine="openpyxl")
            except Exception:
                print(f"⚠️ {file_path} 不是标准xlsx文件，尝试以HTML或CSV读取...")
                try:
                    df = pd.read_html(file_path)[0]
                except Exception:
                    df = pd.read_csv(file_path, encoding="utf-8", engine="python", on_bad_lines="skip")
        elif ext == ".csv":
            df = pd.read_csv(file_path, encoding="utf-8", engine="python", on_bad_lines="skip")
        else:
            print(f"❌ 不支持的文件类型: {file_path}")
            return

        # ===== 示例分析逻辑（你可以换成自己的分析函数）=====
        summary = df.describe(include='all')
        summary.to_excel(os.path.join(output_path, os.path.basename(file_path) + "_summary.xlsx"))
        print(f"✅ 已分析: {file_path}")

    except Exception as e:
        print(f"❌ 无法读取文件 {file_path}: {e}")

# === 设置路径 ===
folder_path = r"D:\工作\特变电工\01_0政策与学习笔记\_全国_20250520\国网区域\新疆\400-交易复盘\现货日清分_系统导出数据\9月1日-9月30日"
output_path = folder_path  # 输出结果放同目录下

for file in os.listdir(folder_path):
    if file.endswith((".xls", ".xlsx", ".csv")):
        analyze_one_file(os.path.join(folder_path, file), output_path)
```
