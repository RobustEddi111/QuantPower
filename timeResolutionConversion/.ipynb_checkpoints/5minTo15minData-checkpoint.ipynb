```python
# 5分钟数据转15分钟数据
```

```python
import numpy as np
import pandas as pd
```

```python
# 读取Excel文件
file_path = 'D:\\Develop\\CodesFiles\\Python\\pythonProject\\TBEADataProcess\\timeResolutionConversion\\测点数据 (6).xlsx'  # 替换为你的Excel文件路径
sheet_name = 'sheet1'  # 替换为你要读取的工作表名称

# 使用read_excel读取数据
data = pd.read_excel(file_path, sheet_name=sheet_name)

# 打印读取的数据
print(data)
```

```python
# 时间转为 datetime 并设为索引
data['时间'] = pd.to_datetime(data['时间'])
data.set_index('时间', inplace=True)
data
```

```python
# 将数据重采样为15分钟数据，使用求和或平均等方法
data_15min = data.resample('15T').mean()/1000  # 或者使用 .mean() 取平均值
```

```python
# 如果需要时间作为一列而不是索引
data_15min = data_15min.reset_index()
data_15min
```

```python
# 写入到 Excel 的 Sheet1 和 Sheet2
with pd.ExcelWriter('output_15min.xlsx', engine='openpyxl') as writer:
    data_15min.to_excel(writer, index=False, sheet_name='Sheet1')
```
