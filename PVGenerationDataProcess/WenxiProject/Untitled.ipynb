```python
import pandas as pd

# 创建示例 DataFrame，假设有 12 个月，每个月有 24 小时的发电量数据
data = {
    '时间': pd.date_range(start='2024-01-01 00:00', periods=24 * 12, freq='H'),
    '发电量': [100 + i * 3 for i in range(24 * 12)]  # 示例数据
}
df = pd.DataFrame(data)

# 将时间列设置为索引
df['时间'] = pd.to_datetime(df['时间'])
df.set_index('时间', inplace=True)

# 生成一个新的 DataFrame 进行透视
df['月'] = df.index.month
df['小时'] = df.index.hour

# 透视表以创建矩阵
matrix = df.pivot(index='小时', columns='月', values='发电量')

# 填充缺失值（如果有的话）
matrix.fillna(0, inplace=True)

# 打印结果
print("矩阵形式的发电量数据:")
print(matrix)
```
